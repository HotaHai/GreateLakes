```python
def plot_corr(df, size=14, title='Correlation'):
    sns.set(font_scale=1.15)
    fig,ax=plt.subplots(figsize=(size,size))
    g = sns.heatmap(df.corr(), annot=True, linewidths=0.01, linecolor="white", fmt="0.2f", cmap='YlGnBu')
    bottom, top = g.get_ylim()
    g.set_ylim(bottom + 0.5, top - 0.5)
    g.set_xticklabels(plt.xticks()[1], rotation=60)
    plt.title(title,fontsize=12)
    ax.tick_params(labelsize=14)
    
plot_corr(cData)
```

### outliers and missing values

```python
cData.skew().sort_values()
```

```python
# Lets remove them:

for col_name in cData.columns[:-1]:
    q1 = cData[col_name].quantile(0.25)
    q3 = cData[col_name].quantile(0.75)
    iqr = q3 - q1
    
    low = q1-1.5*iqr
    high = q3+1.5*iqr
    
    cData.loc[(cData[col_name] < low) | (cData[col_name] > high), col_name] = cData[col_name].median()
    
g = cData.boxplot(figsize=(15,15))
g.set_xticklabels(plt.xticks()[1], rotation=60)
```

### Prepare for model and explore indifivual clusters

```python
x = cData.loc[:, cData.columns != 'strength']
y = cData['strength']
```

```python
stdScale = StandardScaler()
cols = x.columns
x_scaled = stdScale.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=cols)
x_scaled.head()
```

```python
# Split x and y into training and test set in 70:30 ratio
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 10)
```

### KMeans Clustering

```python
kmeans = KMeans(n_jobs=-1, random_state=10)
```

```python
# Selecting k 
clusters=range(1,10)
meanDistortions=[]

for k in clusters:
    model=KMeans(n_clusters=k)
    model.fit(x_scaled)
    prediction=model.predict(x_scaled)
    meanDistortions.append(sum(np.min(cdist(x_scaled, model.cluster_centers_, 'euclidean'), axis=1)) / x_scaled.shape[0])


plt.plot(clusters, meanDistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
```

```python
kmeans = KMeans(n_jobs=-1, n_clusters=3, random_state=10)
kmeans.fit(x_scaled, y)  
y_pred = kmeans.predict(x_scaled)
cData['group'] = y_pred
```

```python
f, axes = plt.subplots(4, 3, figsize=(20, 25))
    
features = np.array(cData.columns[cData.columns != 'group']).reshape(4, 3)
for row in range(4):
    for col in range(3):
        axesplt = axes[row][col]
                
        plot = sns.boxplot(x='group', y=features[row][col], data=cData, ax=axesplt, orient='v', hue='group', palette='Set1')
        plot.set_xlabel('group',fontsize=20)
        plot.set_ylabel(features[row][col],fontsize=20)
```

#### Observations:

##### Body of clustes are overlaping and non of the dimensios are good predictor of the target variable. Though we can see the clusters, we can't find distinct charateristics of these clusters which tells us to break data into different clusters and build separate models for them. 

### Linear Regression

```python
def modelFitAndPredict(model, x_train, y_train, x_test, y_test):    
    model.fit(x_train, y_train)    
    
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)
    
    accuracy_train = round(metrics.r2_score(y_train, y_train_pred),3)
    accuracy_test = round(metrics.r2_score(y_test, y_test_pred),3)
    
    model_result = pd.DataFrame([[accuracy_train],[accuracy_test]], index=['Training', 'Testing'], columns=['Accuracy'])  
        
    return model_result
```

```python
linearModel = LinearRegression(n_jobs=-1)
linearModelResult = modelFitAndPredict(linearModel, x_train, y_train, x_test, y_test)
linearModelResult
```

#### Feature Importance

```python
# summarize feature importance
print ('Scores:')
for i,v in enumerate(linearModel.coef_):
	print('%30s: %.5f' % (x_train.columns[i], v))

plt.bar([x for x in range(len(linearModel.coef_))], linearModel.coef_)
plt.show()
```


#### It tell us water_to_cement_ratio, coarseagg_to_fineagg_ratio adn fineagg_to_cement_ratio are two important features and age, superplastic also has little bit more importance than others.
