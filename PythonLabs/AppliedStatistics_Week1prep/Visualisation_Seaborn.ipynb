Seaborn is a visualisation 
isualization library, it is actually based on Matplotlib.
It was built primarily to provide a high-level interface for drawing
attractive statistical graphics, such as regression plots, box plots,
 and so on.

```
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

Dependencies on pandas, scipy, numpy, matplotlib

```
sns.set(style="darkgrid")
flight = sns.load_dataset("flights")
flight
```

```
sns.relplot(x='passengers', y ="month", data = flight)
#this is for 2 dimension
```

```
sns.relplot(x='passengers', y ="month", hue="year", data = flight)
```

```
sns.catplot(x='passengers', y ="month",  data = flight)
```

```
#the above one is scater plot 
sns.relplot(x='passengers', y ="year", kind="line", data = flight)
#this is for 2 dimension
```

```
tips = sns.load_dataset("tips")
print(tips)
sns.relplot(x="time", y="tip", data=tips, kind="line")
```

```
#plotting for cateogrical data
sns.catplot(x="day", y="total_bill", data=tips )
```

```
sns.catplot(x="day", y="total_bill", data=tips, kind="violin")
```

```
sns.catplot(x="day", y="total_bill", data=tips, kind="boxen")
```

```
# univariate and bivariate distribution
sns.set(style="white")
from scipy import stats
data = np.random.normal(loc=5, scale=10, size =100);
sns.distplot(data);
```

```
x= pd.DataFrame({'Day':[1,2,3,4,5,6,7],'Grocery':[30,80,40,23,51,46,76],'Clothes':[13,40,34,23,54,67,98],'Utilities': [12,32,27,14,58,10,18]})
y= pd.DataFrame({'Day':[8,9,10,11,12,13,14],'Grocery':[30,80,40,23,51,46,76],'Clothes':[13,40,34,23,54,67,98],'Utilities': [12,32,27,14,58,10,18]})
with sns.axes_style("white"):
  sns.jointplot(x=x, y=y, kind = "kde", color="b")
```

```
data = sns.load_dataset("iris")
print(data)
data.describe()
```

```
b= sns.FacetGrid(data, col="species")
b.map(plt.hist, "sepal_length")
```

```
# scatter plot
b= sns.PairGrid(data)
b.map(plt.scatter)
```

```

flight = sns.load_dataset("flights")
b= sns.PairGrid(flight)
b.map(plt.scatter);
```

```
sns.set(style="darkgrid")
flight = sns.load_dataset("flights")
b= sns.PairGrid(flight)
b.map(plt.scatter);
```

```
sns.set(style="dark", color_codes=True)
flight = sns.load_dataset("flights")
sns.boxplot(x="year", y="passengers", data=flight)
```

```
sns.set(style="dark", color_codes=True)
flight = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=flight)
```

```
sns.set(style="white", color_codes=True)
flight = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=flight)
sns.despine(offset=10, trim=True)
```
